```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
w = 1.00
```

```python
def forward(x):
    return x * w

def cost(xs, ys):
    cost = 0
    for x, y in zip(xs, ys):
        y_pred = forward(x)
        cost += (y_pred - y)**2
    return cost / len(xs)

def gradient(xs, ys):
    grad = 0
    for x, y in zip(xs, ys):
        grad += 2 * x * (x * w - y)
    return grad / len(xs)
```

```python
epoch_list = []
cost_list = []
```

```python
for epoch in range(100):
    cost_val = cost(x_data, y_data)
    grad_val = gradient(x_data, y_data)
    w -= 0.01 * grad_val
    print(epoch, w, cost_val)
    epoch_list.append(epoch)
    cost_list.append(cost_val)
print("Predict (after training): )",4,forward(4))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(epoch_list,cost_list,'r')
plt.ylabel('COSS')
plt.xlabel('EPOCH')
plt.show()
```
