```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
w = 1.00
```

```python
def forward(x):
    return x * w

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2

def gradient(x, y):
    return  2 * x * (x * w - y)
```

```python
epoch_list = []
loss_list = []
```

```python
for epoch in range(100):
    for x, y in zip(x_data, y_data):
        grad = gradient(x, y)
        w -= 0.01 * grad
        print(epoch, w, grad)
        l = loss(x, y)
    epoch_list.append(epoch)
    loss_list.append(l)
```

```python
import matplotlib.pyplot as plt
plt.plot(epoch_list,loss_list,'r')
plt.ylabel('COSS')
plt.xlabel('EPOCH')
plt.show()
```
